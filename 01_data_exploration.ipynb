```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def load_who_gho_data(file_path):
    print(f"Current Working Directory: {os.getcwd()}")
    if os.path.exists(file_path):
        print(f"File exists: {file_path}")
    else:
        print(f"File does NOT exist: {file_path}")
    return pd.read_csv(file_path)

def explore_data(df):
    print("First few rows of the dataset:")
    print(df.head())
    
    print("\nSummary statistics:")
    print(df.describe())
    
    print("\nMissing values:")
    print(df.isnull().sum())
    
    return df

def plot_distribution(df, column):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=20)
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

def plot_correlation_heatmap(df):
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()

if __name__ == "__main__":
    file_path = 'F:/HealthMate AI_Project/data/raw/WHO-GHO/WHO.csv'  # Adjust the path as per your structure
    df = load_who_gho_data(file_path)
    df = explore_data(df)
    
    # Example: Plot distribution of numeric data
    plot_distribution(df, 'FactValueNumeric')
    
    # Example: Plot correlation heatmap if applicable
    # Ensure the columns used are numeric and suitable for correlation analysis
    numeric_columns = ['FactValueNumeric', 'FactValueNumericLow', 'FactValueNumericHigh']
    numeric_df = df[numeric_columns]
    plot_correlation_heatmap(numeric_df)
```
